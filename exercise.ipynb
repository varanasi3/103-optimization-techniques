1. Use `matplotlib.pyplot.plot` to produce a plot of the functions 
   $f(x) = e^{-x/10}sin(πx)$ and $g(x) = xe^{−x/3}$ over the interval [0, 10]. 
   Include labels for the x- and y-axes, and a legend explaining which line is which plot.
   Save the plot as a .jpg (“Jpeg”).


```python
import matplotlib.pyplot as plt
import numpy as np

f = lambda x: np.exp(-x/10)*np.sin(np.pi*x)
g = lambda x: x*np.exp(-x/3)

plt.figure(figsize=(16,8))
x = np.linspace(0,10,100)

plt.subplot(2,2,1)
plt.plot(x,f(x),label="F(x)")
plt.xlabel("F(x)",fontsize=20)
plt.ylabel("x",fontsize=20)
plt.grid()
plt.legend()

plt.subplot(2,2,2)
plt.plot(x,g(x),label="G(x)")
plt.xlabel("G(x)",fontsize=20)
plt.ylabel("x",fontsize=20)
plt.grid()
plt.legend()

plt.subplot(2,2,3)
plt.plot(x,f(x),label="G(x),F(x)")
plt.plot(x,g(x))
plt.xlabel("G(x),F(x)",fontsize=20)
plt.ylabel("x",fontsize=20)
plt.grid()
plt.legend()

plt.tight_layout()

plt.savefig('1.jpg')

```

4. Use `matplotlib.pyplot.plot` to produce the surface and contour plot of the function $f(X) = x_1^2x_2^2(1-x_1-x_2).$ Can you identify maxima and minima using cantour plots. For better visualization, make interactive suface plot.

```python
fig= plt.figure(figsize = (12,10))
ax = plt.axes(projection='3d')

x = np.arange(-5,5.1,0.2)
y = np.arange(-5,5.1,0.2)

X, Y = np.meshgrid(x,y)
Z = X**2*Y**2*(1-X-Y)

surf = ax.plot_surface(X,Y,Z, cmap= plt.cm.cool)


ax.contourf(X,Y,Z)

cp = plt.contour(X, Y, Z, cmap='gist_rainbow');
plt.clabel(cp)
```

```python
from mpl_toolkits.mplot3d import Axes3D
f = lambda x1,x2 : x1**3+ x2**3-3*x1*x2

x1 = np.linspace(-5,5,100)
x2 = np.linspace(-5,5,100)


X, Y = np.meshgrid(x1,x2)
F = f(X,Y)

fig = plt.figure(figsize=(12,8))
ax = plt.subplot(projection='3d') 
surf = ax.plot_surface(X,Y,F, cmap= 'twilight_shifted_r')
fig.colorbar(surf,shrink=0.4, aspect=10)
ax.contour(X, Y, F, cmap ='jet')

ax.set_xlabel('x1', labelpad=30)
ax.set_ylabel('x2', labelpad=30)
ax.set_zlabel('f(X)', labelpad=10)

plt.show()
```
