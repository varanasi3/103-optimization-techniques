# 2D Plotting

In Python, the *matplotlib* is the most important package that to make a plott. First import the matplotlib package. In Jupyter notebook, we could show the figure directly within the notebook and also have the interactive operations like pan, zoom in/out, and so on using the magic command - `%matplotlib notebook`. 

````python
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
````

The basic plotting function is *plot(x,y)*. The plot function takes in two lists/arrays, x and y, and produces a visual display of the respective points in x and y.

````python
x = [0,2,9,0]
y = [0,9,7,0]
plt.plot(x,y)
plt.scatter(x,y,color='r')

plt.show()
````

- You can add a legend to your plot by using the *legend* function
- Add a *label* argument in the plot function
- *loc* indicates where to put the legend. Try different numbers and see the location of the legend.
- You can customize the x-range and y-range on the axis using the function *xlim* and *ylim*
- you can use *grid* to add grid to your figure


````python
plt.figure(figsize = (10,6)) 


x = np.linspace(-20,20,1000)
plt.plot(x,x**2,'g--', label='quadratic')
plt.plot(x,x**3,'r*', label='cubic')
plt.plot(x,x**4,'b*', label='cu')
plt.title(f'Plot of variaous polynomials from {x[0]} to {x[-1]}')
plt.xlabel('X-axis', fontsize = 18)
plt.ylabel('Y-axis', fontsize = 18)
plt.legend(loc=2) 
plt.xlim(-10,10)
plt.ylim(-100,100)
plt.grid()
plt.show()

````

## Subplots

- We can create a table of plots on a single figure using the *subplot* function
- The *subplot* function has three inputs: the number of rows, the number of columns of plots, and to which plot all calls to plotting functions should plot.
- Here we try to plot different subplots.


````python

x = np.arange(15)
y = x**2

plt.figure(figsize=(14,8))

plt.subplot(2,3,1) # 2-rows, 3-colums and 1st plot
plt.plot(x,y)
plt.title('Plot: y = $x^2$')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()

# 2-rows, 3-colums and 2st plot
plt.subplot(2,3,2)
plt.scatter(x,y) # scatter plot function
plt.title('Scatter')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()

# 2-rows, 3-colums and 3rd plot
plt.subplot(2,3,3)
plt.bar(x,y) # bar plot function
plt.title('Bar')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()

# 2-rows, 3-colums and 4th plot
plt.subplot(2,3,4)
plt.loglog(x,y) # loglog plot function
plt.title('loglog')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()

# 2-rows, 3-colums and 5th plot
plt.subplot(2,3,5)
plt.semilogx(x,y) # semilogx plot function
plt.title('Semilogx')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()

# 2-rows, 3-colums and 6th plot
plt.subplot(2,3,6)
plt.semilogy(x,y) # semilogy plot function
plt.title('Semilogy')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()


plt.tight_layout()
plt.show()
plt.savefig('subplots.png')
````

# 3D Plotting
Load required toolkit

````python
from mpl_toolkits import mplot3d
plt.style.use('seaborn-poster')
````

Create 3D axes and add data to the axes.

````python
fig= plt.figure(figsize=(10,10))
ax = plt.axes(projection='3d')
plt.show()
````

The `ax = plt.axes(projection=3d)` created a 3D axes object,

````python
# Example

fig = plt.figure(figsize = (15,15))
ax = plt.axes(projection='3d')
ax.grid()
t = np.arange(0, 10*np.pi, np.pi/50)
x = np.sin(t)
y = np.cos(t)

ax.plot3D(x,y,t)
ax.set_title('3D Parametric Plot')

# Set axes label
ax.set_xlabel('x', labelpad=2)
ax.set_ylabel('y', labelpad=2)
ax.set_zlabel('t', labelpad=2)

plt.show()
````

We will make 3D scatter plot.

````python
x = np.random.random(50)
y = np.random.random(50)
z = np.random.random(50)

fig = plt.figure(figsize = (15,15))
ax = plt.axes(projection='3d')
ax.grid()

ax.scatter(x,y,z,marker='*', c = 'g', s = 100,)
ax.set_title('3D scatter Plot')

# Set axes label
ax.set_xlabel('x', labelpad=20)
ax.set_ylabel('y',labelpad=20)
ax.set_zlabel('z', labelpad=20)

plt.show()
````

## Surface Plotting
In surface plotting
- create a mesh using *np.meshgrid*
- Give list/array of x and y values, a mesh is a listing of all the possible (x,y)
- A third array z, can be created such that z(i,j)=f(X(i,j),Y(i,j))


````python
x = [1,2,3,4]
y = [3,4,5]
X, Y = np.meshgrid(x,y)
print(Y)
````

````python
print(X)
````

Plot $f(x,y) = sin(x).cos(y), $ for $-5\le x,y\le 5$ using `plot_surface` function

````python
fig= plt.figure(figsize = (12,10))
ax = plt.axes(projection='3d')

x = np.arange(-5,5.1,0.2)
y = np.arange(-5,5.1,0.2)

X, Y = np.meshgrid(x,y)
Z = np.sin(X)*np.cos(Y)

surf = ax.plot_surface(X,Y,Z, cmap= plt.cm.cool)

# set axes label
ax.set_xlabel('x', labelpad=20)
ax.set_ylabel('y', labelpad=20)
ax.set_zlabel('z', labelpad=20)

fig.colorbar(surf, shrink=0.6, aspect=8) # adding colorbar

plt.show()
````

Surface plotting for the equtaion 
$$z = 4x+6y-2x^2-2xy-2y^2 $$

````python
fig= plt.figure(figsize = (12,10))
ax = plt.axes(projection='3d')

x = np.arange(-10,10,0.2)
y = np.arange(-10,10,0.2)

X, Y = np.meshgrid(x,y)
Z = 4*X+6*Y-2*X**2-2*X*Y-2*y**2

surf = ax.plot_surface(X,Y,Z, cmap= plt.cm.tab20)

# set axes label
ax.set_xlabel('x', labelpad=20)
ax.set_ylabel('y', labelpad=20)
ax.set_zlabel('z', labelpad=20)

fig.colorbar(surf, shrink=0.5, aspect=8)

plt.show()
````

- Note that the surface plot shows different colors for different elevations, yellow for higher and blue for lower,`plt.cm.cividis` function takes care about this.
- There are different color schemes for the surface plot, you can explore about them.
